## 3. Predictive Modeling

```python
# encode varible "Coupons" ---- since it's either offering a coupons (1,000,000) or not in the dataset
dfd = df.drop("Actual Sales", axis=1)
dfd.Coupons = dfd.Coupons.astype(str)
dfd.Coupons = dfd.Coupons.map({"1000000":"yes", "0":"no"}) 
dfd = pd.get_dummies(dfd) # for linear regression
dff = df.drop("Actual Sales", axis=1)
dff.Coupons = dff.Coupons.astype(str)
dff.Coupons = dff.Coupons.map({"1000000":1, "0":0}) # for random forest 
```

### 3.1 Linear Regression

```python
x1 = dfd.drop(["Sales Units", "Display Impressions", "Distribution (ACV)"], axis=1)
y1 = dfd["Sales Units"]
```

```python
m1 = sm.OLS(y1, x1)
results1 = m1.fit()
results1.summary()
```

```python
x1_new = x1.drop(["Search Clicks", "CPI", "Price Reduction Units", "Trade Promotions Units", "Seasonality"], axis=1)
```

```python
m1 = sm.OLS(y1, x1_new)
results = m1.fit()
results.summary()
```

## Regression Diagnostics

#### normality of residuals

```python
# Omni test :
name = ['Chi^2', 'Two-tail probability']
test = sms.omni_normtest(results.resid)
lzip(name, test)
```

```python
# Leverage vs. Normalized Residuals Squared
from statsmodels.graphics.regressionplots import plot_leverage_resid2
fig, ax = plt.subplots(figsize=(8,6))
fig = plot_leverage_resid2(results, ax = ax)
```

```python
# qq plot
sm.qqplot(results.resid)
plt.title("QQ Plot")
plt.show()

import scipy.stats as scipystats
import pylab
scipystats.probplot(results.resid, dist="norm", plot=pylab)
pylab.show()
```

#### Homogeneity of Variance

```python
# Heteroskedasticity tests
# Breush-Pagan test:
name = ['Lagrange multiplier statistic', 'p-value', 
        'f-value', 'f p-value']
test = sms.het_breushpagan(results.resid, results.model.exog)
lzip(name, test)
```

#### Linearity

```python
# Linearity
import statsmodels
import statsmodels.stats.api as sms
name = ['t value', 'p value']
test = sms.linear_harvey_collier(results)
lzip(name, test)
```

#### Multicollinearity

```python
np.linalg.cond(results.model.exog)
```

#### Metrics Summary

```python
# define functions for model validation
# definte RMSE function
from sklearn.metrics import mean_squared_error
from math import sqrt

def rmse(x, y):
    return sqrt(mean_squared_error(x, y))

# definte MAPE function
def mape(x, y): 
    return np.mean(np.abs((x - y) / x)) * 100  
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression().fit(x1, y1)
y1_predicted = linreg.predict(x1)
print("Linear Regression Model 1 Metrics:")
print("R_square" , ":" , linreg.score(x1, y1))
print("AIC : 701.6")
print("BIC : 714.2")
print("RMSE", ":", rmse(y1, y1_predicted))
print("MAPE", ":", mape(y1, y1_predicted))
```

```python
linreg2 = LinearRegression().fit(x1_new, y1)
y1_predicted = linreg2.predict(x1_new)
print("Linear Regression Model 2 Metrics:")
print("R_square" , ":" , linreg2.score(x1_new, y1))
print("AIC : 699.8")
print("BIC : 705.4")
print("RMSE", ":", rmse(y1, y1_predicted))
print("MAPE", ":", mape(y1, y1_predicted))
```

### 3.2 Random Forest

```python
x2 = dff.drop("Sales Units", axis=1)
y2 = dff["Sales Units"]
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=30)
rfreg = rf.fit(x2,y2)
```

```python
# feature importance
features = x2.columns
importances = rfreg.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(8,10))
plt.title('Feature Importances', fontsize=20)
plt.barh(range(len(indices)), importances[indices], color='pink', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.xlabel('Relative Importance')
```

```python
y2_predicted = rfreg.predict(x2)
print("R_square" , ":" , rfreg.score(x2, y2))
print("RMSE", ":", rmse(y2, y2_predicted))
print("MAPE", ":", mape(y2, y2_predicted))
```
